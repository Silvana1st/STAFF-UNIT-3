### 1. Create a virtual environment for the Pycharm Tutorial

We will use the sprint virtual environment for this tutorial so that a single environment can be used for this entire course, however let's look back at how the environment was created.

```bash
$ conda create create --name sprint python=3.7
$ conda activate sprint
```

These commands created the virtual environment, installed Python 3 into it and finally started the environment. I have also linked to the [conda cheat-sheet](https://docs.conda.io/projects/conda/en/4.6.0/_downloads/52a95608c49671267e40c689e0bc00ca/conda-cheatsheet.pdf)

### 2. Configure PyCharm project interpreter
We needs to tell PyCharm which Python interpreter to use for debugging our code. 

This website tells us how to [Configure the Python Interpreter](https://www.jetbrains.com/help/pycharm/conda-support-creating-conda-virtual-environment.html).

We need to identify the Python executable. Run the following cell.

````python
! which python
````

This is the Python executable you will specify in [Configure the Python Interpreter](https://www.jetbrains.com/help/pycharm/conda-support-creating-conda-virtual-environment.html), section 3, bullet point 1.
